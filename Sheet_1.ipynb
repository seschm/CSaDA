```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
def simulation(N):

  boxes = 0
  num_dif_coup = 0

  while num_dif_coup < N:
    p_new = (N-num_dif_coup)/N
    coupon = np.random.choice(['new','old'],p=[p_new,1-p_new])
    if coupon == 'new':
      boxes += 1
      num_dif_coup += 1
    else:
      boxes += 1

  return boxes
```

```python
average_box_num = []

for i in range(1,51):

  counter = []
  for j in range(0,100):
    box_num = simulation(i)
    counter.append(box_num)

  average_box_num.append(np.mean(counter))
```

```python
plt.scatter(np.linspace(1,50,50),average_box_num, s=5)
plt.grid(True, linestyle='--')
plt.xlabel(r'number of different coupons $N$')
plt.ylabel('number of expected boxes')
plt.title('expected number of boxes for 100 simulations')
```
