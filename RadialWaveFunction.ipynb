```python
# Load parameters for Rubidium87
atom = Rubidium87()
n = 82 # principal quantum number
l = 2 # D state
j = 3/2 # J = 3/2
step = 0.001

a1, b1 = atom.radialWavefunction(
    l,
    0.5,
    j,
    atom.getEnergy(n, l, j) / 27.211,
    atom.alphaC ** (1 / 3.0),
    16000,
    step,
)
```
